````python
from sympy.solvers import solve
from sympy import Symbol, Eq
import sys
from sympy.solvers.solveset import linsolve

a = Symbol('a')
b = Symbol('b')
c = Symbol('c')

d = Symbol('d')
e = Symbol('e')
f = Symbol('f')

g = Symbol('g')
h = Symbol('h')

a, b, c = 0.2, 0.3, 0.5
d, e = 0.6, 0.4 
f = 0.6
g, h = 0.2, 0.3

l = [Symbol('l%d' %i) for i in range(10)]
sol=linsolve(
    [
        l[1] + f*l[2] - l[2],
        a*l[2] + h*l[4] - l[3],
        b*l[2] + g*l[3] - l[4],
        c*l[2] - l[5],
        d*l[2] - l[6],
        e*l[2] - l[7],
    ], *[l[x] for x in range(2, 8)]
)
ss = list(sol)[0]
print("Number of solutions:", len(sol))


leff = [None for _ in range(10)]
leff[1] = l[1]
for i, ass in zip(range(2, 8), ss):
    leff[i] = ass
    print(f"l{i}: {leff[i]}")

S = [None, 0.003, 0.001, 0.01, 0.04, 0.1, 0.13, 0.15]
ro = [None for _ in range(10)]
Qk = [None for _ in range(10)]

L = 1.0

for i in range(1, 8):
    ro[i] = S[i] * leff[i]
    print(f"ro[{i}]: {ro[i].subs(l[1], L)}")  # Utilization
    Qk[i] = ro[i] / (1 - ro[1])
R = sum(xx for xx in Qk[1:8])/l[1]
print(R)
print(R.subs(l[1], L))
print(leff[1:8])
````

````python
import pdq
pdq.Init("0036477241")
pdq.CreateOpen("wn", 1.0)
for i in range(1, 8):
    pdq.CreateNode(f"n{i}", pdq.CEN, pdq.FCFS)
for i in range(1, 8):
    pdq.SetDemand(f"n{i}", "wn", S[i] * float(leff[i].subs(l[1], 1.0)))
pdq.Solve(pdq.CANON)
# with sys.stdout as sys.stderr:
pdq.Report()
R.subs(l[1], 1)
````
