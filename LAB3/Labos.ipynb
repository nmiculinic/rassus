````python
from sympy.solvers import solve
from sympy import Symbol, Eq, simplify
import sys
from sympy.solvers.solveset import linsolve

a = Symbol('a')
b = Symbol('b')
c = Symbol('c')

d = Symbol('d')
e = Symbol('e')
f = Symbol('f')

g = Symbol('g')
h = Symbol('h')

a, b, c = 0.2, 0.3, 0.5
d, e = 0.6, 0.4 
f = 0.6
g, h = 0.2, 0.3

S = [None, 0.003, 0.001, 0.01, 0.04, 0.1, 0.13, 0.15]
L = Symbol("L")


l = [Symbol('l%d' %i) for i in range(10)]

sol=linsolve(
    [
        L - l[1],
        l[1] + f*((1 - g)*l[3] + (1 - h)*l[4] + l[5]) - l[2],
        a*l[2] + h*l[4] - l[3],
        b*l[2] + g*l[3] - l[4],
        c*l[2] - l[5],
        d*(1 - f)*((1 - g)*l[3] + (1 - h)*l[4] + l[5]) - l[6],
        e*(1 - f)*((1 - g)*l[3] + (1 - h)*l[4] + l[5]) - l[7],
    ], *[l[x] for x in range(1, 8)]
)
ss = list(sol)[0]
print("Number of solutions:", len(sol))


leff = [None for _ in range(10)]
leff[1] = l[1]
for i, ass in zip(range(1, 8), ss):
    leff[i] = ass
    print(f"l{i}: {leff[i]}")

ro = [None for _ in range(10)]
Qk = [None for _ in range(10)]


for i in range(1, 8):
    ro[i] = S[i] * leff[i]
    print(f"ro[{i}]: {ro[i]}")  # Utilization
    Qk[i] = ro[i] / (1 - ro[i])
for i in range(1, 8):
    print(f"S[{i}]: {S[i]}")
for i in range(1, 8):
    print(f"leff[{i}]: {leff[i]}")

    
R = simplify(sum(xx for xx in Qk[1:8])/L)
# print("R", R)
print(R.subs(L, 1))
````

````python
import seaborn as sns
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 7.5)  # 8 je max
y = np.array([R.evalf(subs={L:xx}) for xx in x])
plt.plot(x, y)
plt.xlabel("Lambda")
plt.ylabel("Residence time");
````

````python
import pdq
pdq.Init("lab2.1")
pdq.CreateOpen("wn", 1.0)
for i in range(1, 8):
    pdq.CreateNode(f"n{i}", pdq.CEN, pdq.FCFS)
for i in range(1, 8):
    pdq.SetDemand(f"n{i}", "wn", S[i] * float(leff[i].subs(L, 1.0)))
pdq.Solve(pdq.CANON)
# with sys.stdout as sys.stderr:
pdq.Report()
````
